3) Obtener el TOP de las 3 propiedades más caras, para cada tipo de propiedad.

```python
import pandas as pd
house_apartments = pd.read_excel('../Dataset_Inmuebles.xlsx', sheet_name = 'Inputs_1')
df_inmuebles = pd.read_csv('../inmuebles_modificado.csv')

conjunto_house_apartments = pd.merge(house_apartments,df_inmuebles, on='ID')

conjunto_h_ap_ordenado = conjunto_house_apartments.sort_values(['property_type','price_usd'], ascending=[True,False])

top_3_x_categoria = conjunto_h_ap_ordenado.groupby('property_type').head(3)
top_3_x_categoria
```
