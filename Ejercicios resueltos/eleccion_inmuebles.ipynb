Un inversionista posee un presupuesto de $800.000 y desea llevar a cabo las siguientes inversiones:

1. Adquirir la mayor cantidad posible de inmuebles priorizando la superficie total
2. Desea adquirir 3 inmuebles priorizando la superficie total y con expensas bajas

```python
import pandas as pd
inmuebles = pd.read_csv('../precio_superficie_expensas.csv')
inmuebles
```

Ejercicio 1

```python
inmuebles = inmuebles.sort_values(by='surface_total', ascending = False)

cantidad_inmuebles = 0
precio_total = 0
inversion = 800000

for index, fila in inmuebles.iterrows():
    if (inversion - (fila['price_usd'] + fila['expenses'])) > 0:
        precio_total = precio_total + fila['price_usd'] + fila['expenses']
        inversion = inversion - (fila['price_usd'] + fila['expenses'])
        cantidad_inmuebles+=1
        
        indice = inmuebles.loc[index]
        print(indice)
        print('-----------------------------')
        
        
print('-----------------------------')
print('precio total de la inversion:', precio_total)
print('cantidad inmuebles adquiridos:', cantidad_inmuebles)

    


    
```
