6) ¿Cuál es el barrio con mayor costo promedio por metro cuadrado cubierto?

```python
import pandas as pd
df_inmuebles = pd.read_csv('../df_inmuebles_completo.csv')
df_inmuebles
```

```python
df_inmuebles['metro_cuadrado'] = df_inmuebles['price_usd'] / df_inmuebles['surface_covered']

metro_cuadrado_en_barrios = df_inmuebles.groupby('barrio')['metro_cuadrado'].mean()
metro_cuadrado_en_barrios
```
